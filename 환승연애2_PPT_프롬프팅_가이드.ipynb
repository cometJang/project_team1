```
3개 케이스를 정의해줘:

**Case 1: 재결합 성공**
- 커플: 이나연 ♥ 남희두
- MBTI: ESFP x INTP
- 속마음 문자: 6 vs 7 (균형!)
- 연애 기간: 31개월 | 이별 5개월
- 결과: 재결합 성공 💕

**Case 2: 환승 성공**
- 커플: 성해은 ♥ 정현규
- MBTI: INFP x ESTJ
- 속마음 문자: 4 vs 5 (균형적)
- 이전 연애:
  - 해은(X규민): 76개월 | 이별 15개월
  - 현규(X나언): 14개월 | 이별 30개월
- 결과: 환승 성공 ✨

**Case 3: 재결합 실패**
- 커플: 성해은 x 정규민
- MBTI: INFP x ENTJ
- 속마음 문자: 10 vs 0 (일방적)
- 연애 기간: 76개월 | 이별 15개월
- 결과: 재결합 실패 💔
```

```
감정 심층 분석 슬라이드를 듀얼 차트로 만들어줘:

**좌측 차트: 무관 강도 비율 (Indifference)**
- 정의: X(전 연인) 대화 중 무관심/회피적 반응의 비율
- 분석 키워드: "상관없어", "별로", "그냥"
- 데이터:
  - 이나연: 38.2%
  - 남희두: 33.8%
  - 성해은: 40.9% (1위) ⭐
  - 정규민: 29.4%

**우측 차트: 긍정 감정도 (Positive Sentiment)**
- 정의: 전체 대화 맥락에서 긍정/기대 표현의 빈도
- 분석 키워드: "좋아", "기대돼", "행복해"
- 데이터:
  - 이나연: 35.3%
  - 남희두: 22.2%
  - 성해은: 40.2% (1위) ⭐
  - 정규민: 20.6%

**인사이트 박스:**
- 제목: "높은 긍정도의 역설 (일방적 미련)"
- 내용: 성해은의 긍정 감정도(40.2%, 1위)는 X(규민)에 대한 높은 관심과 기대감을 의미. (동시에 무관 강도 40.9%로 1위). 이는 상호작용이 결여된 일방적인 미련의 신호였으며, 결과적으로 관계의 부담이 되어 재결합 실패의 원인이 되었습니다.
- 참고: 무관 강도는 해은의 "그냥" 말버릇 영향 가능성 있음

**상단 우측 배지:** "Gemini 2.0 Flash로 807개 대화의 감정 톤 분석"
```
